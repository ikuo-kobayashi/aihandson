```python
# １．csvファイルの読み込み＆冒頭部分の確認【4】【1】
# （読み込みたいCSVファイルは "../input/titanic3.csv" ）

# csvファイルを読み込み「df」に代入

df = pd.read_csv("../input/titanic3.csv")

# 冒頭部分の確認
df.head()
```

```python
# ２．指定した（範囲の）行、列の抽出【21】【22】

# dfから「name」と「age」と「fare」の列だけを抽出し「df2_1」に代入
df2_1 = df.loc[:,['name','age','fare']]

# dfのインデックス番号「100～150」まで かつ　列名「name」～「cabin」までを抽出し「df2_2」に代入
df2_2 = df.loc[100:150,'name':'cabin']
```

```python
# ３．csvファイルへの出力【24】

# df2_2を "../output/review_3.csv" に保存する。ただしインデックス番号は出力しないこと。
df2_2.to_csv("../output/review_3.csv", index=False)
```

```python
# ４．条件に合致した行の抽出【27】【29】

# dfから「age」が「20以上50以下」かつ「home.dest」に「New York」が含まれる行のみを抽出し「df4_1」に代入
df4_1 = df.loc[((df['age']>=20)&(df['age']<=50))&df['home.dest'].str.contains('New York')]

# dfから「name」～「cabin」までの列を抽出し「df4_2」に代入
df4_2 = df.loc[:,'name':'cabin']
```

```python
df4_1.head()
```

```python
df4_2.head()
```

```python
# ５．条件に合致した箇所の値の書き換え【34】

# dfをdf5にコピーする
df5 = df.copy()

# df5に対して「sex」が「female」かつ「age」が「20未満」の人については、「fare」を0にして「fare」に上書きする
df5.loc[(df5['sex']=='female')&(df5['age']<20),'fare'] = 0

df5.head()
```

```python
# ６．特定の列に対する一律な四則演算【36】

# dfをdf6にコピーする
df6 = df.copy()

# df6に対して「fare」を1.1倍して「fare」に上書きする
df6['fare'] = df6['fare']*1.1
```

```python
# ７．複数列にわたる文字列の結合と新規列への代入【40】

# dfをdf7にコピーする
df7 = df.copy()

# df7に対して、「sex」と「age」を間に「_」を挟んで文字列として結合して、「sex-age」という新しい列に代入する
df7['sex_age'] = df7['sex'] + '_' + df7['age'].astype(str)

df7.head()
```

```python
# ８．列名の変更【44】

# dfをdf8にコピーする
df8 = df.copy()

# 列名「pclass」を「rank」に、「home.dest」を「OD」に変更する
df8 = df.rename(columns={'pclass':'rank','home.dest':'OD'})
df8.head()
```

```python
# ９．欠損値の補間【47】

# dfをdf9にコピーする
df9 = df.copy()

# df9の「age」、「body」をそれぞれの平均値で補間する
df9['age'] = df9['age'].fillna(df9['age'].mean())
df9['body'] = df9['body'].fillna(df9['age'].mean())

# まとめて実施する場合
# df9[['age','body']] = df9[['age','body']].fillna(df9[['age','body']].mean())

df9
```

```python
# １０．文字列の置換【56】

# dfをdf10にコピーする
df10 = df.copy()

# df10の「name」列について「Miss」または「Mrs」を「Ms」に置換する
df10['name'] = df10['name'].replace('Miss','Ms',regex=True)
df10['name'] = df10['name'].replace('Mrs','Ms',regex=True)

# まとめて実施する場合
# df10['name'] = df10['name'].replace(['Miss','Mrs'],'Ms',regex=True)
df10
```

```python
# １１．データフレームの結合【59】【65】

# ふたつのデータフレームを読み込み
df11_1 = pd.read_csv('../input/data1.csv')
df11_2 = pd.read_csv('../input/data1_2.csv')

# df11_1とdf11_2をnameで左結合してdf11_3に代入
df11_3 = pd.merge(df11_1,df11_2,on='name',how='left')

# df11_1とdf11_2を行方向に単純結合してdf11_4に代入
df11_4 = pd.concat([df11_1,df11_2],axis=0)
```

```python
# １２．統計値の算出【66】【68】【71】

# データフレームを読み込み
df12_1 = pd.read_csv('../input/data1.csv')

# df12_1の「English」「Mathematics」「History」のそれぞれの平均値を算出してdf12_2に代入
df12_2 = df12_1[['English','Mathematics','History']].mean()

# df12_1の「class」でグルーピングして、グループごとの「English」「Mathematics」「History」のそれぞれの平均値を算出してdf12_3に代入
df12_3 = df12_1.groupby('class')[['English','Mathematics','History']].mean()

# df12_1の各種統計値を一括算出してdf12_4に代入
df12_4 = df12_1.describe()
```
