# タイタニック号の生存者予測 (90 - 100相当)  【小林オリジナル】

```python
# ０．データフレームの準備/確認
df_copy = df.copy()
df_copy
```

```python
# １．各列の欠損状況を確認する　【46】
df_copy.isnull().sum()
```

```python
# ２．明らかに不要そうな列、リーク情報になりそうな列を削除する（任意）　【41】
df_copy2 = df_copy.drop(['name','ticket','boat','home.dest'], axis=1)
df_copy2
```

```python
# ３．ラベルエンコーディングを実施する（文字列のままではAI学習できないため）　【80】
#（from sklearn.preprocessing import OrdinalEncoderを使用）
from sklearn.preprocessing import OrdinalEncoder
import numpy as np
oe = OrdinalEncoder()
df_copy3 = oe.fit_transform(df_copy2)
df_copy4 = pd.DataFrame(df_copy3, columns=df_copy2.columns)
df_copy4
```

```python
# ４．各列の欠損部分を何らかの方法（平均値等）で補間する　【47】
df_copy5 = df_copy4.fillna(df_copy4.mean())
df_copy5
```

```python
# ５．各列の欠損状況を再確認する　【46】
df_copy5.isnull().sum()
```

```python
# ６．学習データとテストデータに分割する（概ね学習データ７割、テストデータ３割が目安）　【42】【50】
df_train = df_copy5.sample(frac=0.7)
df_test = df_copy5.drop(df_train.index, axis=0) # 元のデータフレームから学習データの行（index）を削除することで生成している
```

```python
df_train
```

```python
df_test
```

```python
# ７．学習データとテストデータそれぞれについて説明変数と目的変数に分割する　【41】

df_train_features = df_train.drop('survived',axis=1)
df_train_target = df_train['survived']

df_test_features = df_test.drop('survived',axis=1)
df_test_target = df_test['survived']
```

```python
df_train_features
```

```python
df_train_features
```

```python
df_test_features
```

```python
df_test_target
```

```python
# ８．AI予測モデルを構築する（学習データを使って学習させる）　【new】
#(from sklearn.ensemble import RandomForestClassifierを使用)

from sklearn.ensemble import RandomForestClassifier

rc = RandomForestClassifier()

model = rc.fit(df_train_features, df_train_target)
```

```python
# ９．AI予測モデルを使って予測する（テストデータを使って予測させる）　 【new】

pred = model.predict(df_test_features)
pred
```

```python
df_test_target.values
```

```python
#１０．AI予測モデルの性能を評価する（テストデータを使って予測結果と答えを比較する）　 【new】
#(from sklearn.metrics import accuracy_scoreを使用)

from sklearn.metrics import accuracy_score

accuracy_score(pred, df_test_target)
```

```python
# 参考として混同行列の表示方法
# (from sklearn.metrics import confusion_matrixを使用)
from sklearn.metrics import confusion_matrix

confusion_matrix(pred, df_test_target)
```

```python
# １１．重要だった説明変数を確認してグラフ化する　【38】【84】

x = df_train_features.columns
y = model.feature_importances_
df_plot = pd.DataFrame()
df_plot['features'] = x
df_plot['importances'] = y
df_plot.plot.bar(x='features')
```

# 番外編 autogluonによるAI予測モデル・評価

```python
# 不要・リーク説明変数の削除（５．相当）
df = df.drop(['name','ticket','boat','home.dest'],axis=1)

# 学習データとテストデータに分割（６．相当）
df_train = df.sample(frac=0.7)
df_test = df.drop(df_train.index, axis=0)
```

```python
# autogluonによる予測モデル作成・精度比較

# (from autogluon.tabular import TabularDataset, TabularPredictorを利用)
from autogluon.tabular import TabularDataset, TabularPredictor

# autogluonインスタンスの生成（labelには目的変数の列を指定）
tp = TabularPredictor(label='survived')

# インスタンスのfitに学習データを渡して、AI予測モデルの作成（少し時間がかる）
model = tp.fit(df_train)

# 結果の一覧
model.leaderboard()
```

```python
# テストデータによる評価
model.evaluate(df_test)
```

```python
# 重要説明変数の確認
importance = model.feature_importance(df_test)
importance['importance'].plot.bar()
```
